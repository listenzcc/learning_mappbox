```python
import time
import numpy as np
import pandas as pd
```

```python
url_gj = 'http://eq-tj.ac.cn/index/gjdzml/'
url_tj = 'http://eq-tj.ac.cn/index/tjdzml/'
```

```python
pds = pd.read_html(url_gj)
pds[0]
```

```python
s = pds[0].loc[0]['发震时间']
time.mktime(time.strptime(s, '%Y-%m-%d %H:%M:%S'))
```

```python
df = pd.DataFrame(columns=('time', 'gname', 'loc', 'dep', 'str'))
print('Example:')
print(pds[0].loc[0])

def parse_time(t):
    return time.mktime(time.strptime(t, '%Y-%m-%d %H:%M:%S'))

def parse_pos(s):
    for e in ['北', '南', '西', '东', '经', '纬', '度']:
        s = s.replace(e, '')
    return float(s)

for p in pds:
    p.columns = pds[0].columns
    for j in p.index:
        e = p.loc[j]
        if str(e['发震时间']) in ['nan', '发震时间']:
            continue
        df = df.append([{'time': parse_time(e['发震时间']),
                         'gname': e['震中位置'],
                         'loc': [parse_pos(e['震中纬度']), parse_pos(e['震中经度'])],
                         'dep': float(e['震源深度(Km)']),
                         'str': float(e['震级(M)']),
                        }],
                       ignore_index=True)
df.to_json('earth_quake.json')
df
```
